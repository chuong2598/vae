```python

import torch
from matplotlib import pyplot as plt
import numpy as np
import torchvision
from tqdm.auto import tqdm

device = "cuda" if torch.cuda.is_available() else "cpu"
print(torch.cuda.get_device_name(0))
```

```python

# # Data loaders
trainloader = torch.utils.data.DataLoader(
    torchvision.datasets.CIFAR10(root='./data',train=True,download=True, transform=torchvision.transforms.ToTensor()),
    batch_size=128,shuffle=True)
testloader = torch.utils.data.DataLoader(
    torchvision.datasets.CIFAR10(root='./data',train=False,download=True, transform=torchvision.transforms.ToTensor()),
    batch_size=128,shuffle=True)


# Data loaders
# trainloader = torch.utils.data.DataLoader(
#     torchvision.datasets.MNIST(root='./data',train=True,download=True, transform=torchvision.transforms.ToTensor()),
#     batch_size=128,shuffle=True)
# testloader = torch.utils.data.DataLoader(
#     torchvision.datasets.MNIST(root='./data',train=False,download=True, transform=torchvision.transforms.ToTensor()),
#     batch_size=128,shuffle=True)

```

```python


class VAE(torch.nn.Module):
    
    def __init__(self, image_shape, latent_dim, encoder_hidden_dim, decoder_hidden_dim):
        super(VAE, self).__init__()
        
        channel, height, width = image_shape
        
        self.encoder_hidden_dim = encoder_hidden_dim
        self.latent_dim = latent_dim
        self.image_shape = image_shape
        
        self.encoder = torch.nn.Sequential(
            torch.nn.Linear(in_features=channel*height*width, out_features=encoder_hidden_dim),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(in_features=encoder_hidden_dim, out_features=encoder_hidden_dim),
            torch.nn.LayerNorm(encoder_hidden_dim),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(in_features=encoder_hidden_dim, out_features=encoder_hidden_dim),
            torch.nn.LayerNorm(encoder_hidden_dim),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(in_features=encoder_hidden_dim, out_features=encoder_hidden_dim),
            torch.nn.LayerNorm(encoder_hidden_dim),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(in_features=encoder_hidden_dim, out_features=encoder_hidden_dim),
            torch.nn.LayerNorm(encoder_hidden_dim),
            torch.nn.ReLU(inplace=True),
        )
        
        self.mu_encoder = torch.nn.Linear(in_features=encoder_hidden_dim, out_features=latent_dim)
        self.logvar_encoder = torch.nn.Linear(in_features=encoder_hidden_dim, out_features=latent_dim)
        
        self.decoder = torch.nn.Sequential(
            torch.nn.Linear(in_features=latent_dim, out_features=decoder_hidden_dim),
            torch.nn.LayerNorm(encoder_hidden_dim),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(in_features=decoder_hidden_dim, out_features=decoder_hidden_dim),
            torch.nn.LayerNorm(decoder_hidden_dim),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(in_features=decoder_hidden_dim, out_features=decoder_hidden_dim),
            torch.nn.LayerNorm(decoder_hidden_dim),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(in_features=decoder_hidden_dim, out_features=decoder_hidden_dim),
            torch.nn.LayerNorm(encoder_hidden_dim),
            torch.nn.ReLU(inplace=True),
            torch.nn.Linear(in_features=decoder_hidden_dim, out_features=decoder_hidden_dim),
            torch.nn.LayerNorm(encoder_hidden_dim),
            torch.nn.ReLU(inplace=True),
        )

        self.mu_decoder = torch.nn.Linear(in_features=decoder_hidden_dim, out_features=channel*height*width)
        self.logvar_decoder = torch.nn.Linear(in_features=decoder_hidden_dim, out_features=channel*height*width)
        
    def reparameterize(self, mu, logvar, device="cuda"):
        e = torch.randn(mu.shape[1], device=device) 
        var = torch.exp(logvar)
        std = torch.sqrt(var)
        z = e*std + mu
        return z

    def sample(self, nb_samples, device="cuda"):
        z = torch.randn(nb_samples, self.latent_dim, device=device)
        decoder_output = self.decoder(z)
        mu_decoder = self.mu_decoder(decoder_output)
        logvar_decoder = self.logvar_decoder(decoder_output)
        generated_samples = self.reparameterize(mu_decoder, logvar_decoder)
        generated_samples = torch.reshape(generated_samples, (nb_samples, image_shape[0], image_shape[1], image_shape[2]))
        return generated_samples
        
    def forward(self, images):
        batch, channel, height, width = images.shape
        images_flatten = images.reshape(batch, -1)
        encoder_output = self.encoder(images_flatten)
        
        mu_encoder = self.mu_encoder(encoder_output)
        logvar_encoder = self.logvar_encoder(encoder_output)
        z = self.reparameterize(mu_encoder, logvar_encoder)
        
        decoder_output = self.decoder(z)
        mu_decoder = self.mu_decoder(decoder_output)
        logvar_decoder = self.logvar_decoder(decoder_output)

        pred = self.reparameterize(mu_decoder, logvar_decoder)
        # print(pred.shape)
        pred = torch.reshape(pred, images.shape)
        # print(pred.shape)
        return pred, mu_encoder, logvar_encoder

```

```python

def get_kl_loss(mu, logvar):
    var = torch.exp(logvar)
    kl_loss = -0.5 * (torch.sum(logvar+1) - torch.sum(mu*mu) - torch.sum(var))
    return kl_loss
```

```python


image_shape=(3,32,32)
laten_dim=100
encoder_hidden_dim = 2048
# encoder_kernel_size = 5 
decoder_hidden_dim = 2048 
# decoder_kernel_size = 5
epochs = 1000
lr = 0.0001

vae = VAE(image_shape=image_shape, latent_dim=laten_dim, encoder_hidden_dim=encoder_hidden_dim, decoder_hidden_dim=decoder_hidden_dim)
vae.to(device=device)
optim = torch.optim.Adam(params=vae.parameters(), lr=lr)
mse_loss = torch.nn.MSELoss(reduction='sum')
# bce_loss = torch.nn.functional.binary_cross_entropy
kl_loss = get_kl_loss

```

```python
optim.param_groups[0]['lr'] = 0.00002
```

```python

train_loss = []
val_loss = []
for epoch in range(epochs):
    # Train
    vae.train()
    epoch_train_loss = []
    for original_images, labels in tqdm(trainloader):
        original_images = original_images.to(device=device)
        original_images = original_images 
        reconstructed_images, mu, logvar = vae(original_images)
        # loss = (bce_loss(reconstructed_images, original_images, reduction='sum') + kl_loss(mu, logvar)) / len(original_images)
        loss = (mse_loss(reconstructed_images, original_images) + kl_loss(mu, logvar)) / len(original_images)
        optim.zero_grad()
        loss.backward()
        optim.step()
        epoch_train_loss.append(loss.item())
    train_loss.append(np.mean(epoch_train_loss))

    # vae.eval()
    # with torch.no_grad():
        # Eval
    epoch_val_loss = []
    for original_images, labels in tqdm(testloader):
        original_images = original_images.to(device=device)
        reconstructed_images, mu, logvar = vae(original_images)
        # loss = (bce_loss(reconstructed_images, original_images, reduction='sum') + kl_loss(mu, logvar)) / len(original_images)
        loss = (mse_loss(reconstructed_images, original_images) + kl_loss(mu, logvar)) / len(original_images)
        # loss = (mse_loss(reconstructed_images, original_images) + kl_loss(mu, logvar)) / len(original_images)
        optim.zero_grad()
        loss.backward()
        optim.step()
        epoch_val_loss.append(loss.item())
    val_loss.append(np.mean(epoch_val_loss))
    print(f"Finish epoch {epoch}. Training loss = {np.mean(epoch_train_loss)}, Validation loss = {np.mean(epoch_val_loss)}")

    # Sample
    nb_samples = 100
    generated_samples = vae.sample(nb_samples=nb_samples)
    grid_images = torchvision.utils.make_grid(tensor=generated_samples, nrow=int(np.sqrt(nb_samples)))
    grid_images = grid_images.permute(1,2,0).detach().cpu().numpy() 
    plt.figure(figsize=(7.5,7.5))
    plt.title("Sample images")
    plt.imshow(grid_images)
    plt.show()

    print("-"*100)
```

```python
check_point = {
    "model_state_dict": vae.state_dict(),
    "optim_state_dict": optim.state_dict()
}
torch.save(check_point, "./vae.pt")
```
